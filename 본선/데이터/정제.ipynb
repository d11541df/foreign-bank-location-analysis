```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 대용량 데이터셋 생성
np.random.seed(42)
months = pd.date_range(start='2021-01', end='2024-12', freq='M')
data = {'Date': months, 'Value': np.random.randint(100, 200, size=len(months))}
df = pd.DataFrame(data)

# 연도별로 데이터 정렬
df['Year'] = df['Date'].dt.year
df.sort_values(by=['Year', 'Date'], inplace=True)

# 시각화
plt.figure(figsize=(12, 6))

# 각 연도의 데이터를 연결된 선 그래프로 플로팅
for year in range(2021, 2025):
    subset = df[df['Year'] == year]
    plt.plot(subset['Date'], subset['Value'], marker='o', label=str(year), linestyle='-')

plt.title('대용량 데이터셋 연도별 월 단위 연결된 선 그래프')
plt.xlabel('월')
plt.ylabel('값')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# 가상의 실제값과 예측값 생성
y_true = np.array([100, 150, 200, 250, 300])
y_pred = np.array([110, 140, 210, 240, 290])

# R2 score 계산
r2 = r2_score(y_true, y_pred)

# RMSE 계산
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

# 결과 출력
print(f'R2 Score: {r2:.4f}')
print(f'RMSE: {rmse:.2f}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# 대용량 데이터셋 생성
np.random.seed(42)
months = pd.date_range(start='2021-01', end='2024-12', freq='M')
data = {'Date': months, 'Value': np.random.randint(150, 250, size=len(months))}
df = pd.DataFrame(data)

# 2022, 2023, 2024 값 높이기
df.loc[df['Date'].dt.year.isin([2022, 2023, 2024]), 'Value'] += 80

# 연도별로 데이터 정렬
df['Year'] = df['Date'].dt.year
df.sort_values(by=['Year', 'Date'], inplace=True)

# 한국어 폰트 설정
plt.rcParams['font.family'] = 'NanumBarunGothic'

# 시각화
plt.figure(figsize=(12, 6))

# 각 연도의 데이터를 연결된 선 그래프로 플로팅
for year in range(2021, 2025):
    subset = df[df['Year'] == year]
    plt.plot(subset['Date'], subset['Value'], marker='o', label=str(year), linestyle='-')

plt.title('2021년부터 2024년까지 대용량 데이터 연도별 월 단위 선 그래프')
plt.xlabel('월')
plt.ylabel('값')
plt.legend()
plt.grid(True)

# Y축에 지수 표기법 추가
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True, useOffset=True))
plt.gca().tick_params(axis='y', which='both', labelleft=True)

plt.show()
```
