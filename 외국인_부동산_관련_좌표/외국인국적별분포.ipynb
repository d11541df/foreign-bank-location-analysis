```python
import pandas as pd
```

```python
data = pd.read_csv("foreign_country.csv", encoding='utf-8')
```

```python
data
```

```python
data.info()
```

```python
columns_to_keep = [col for col in data.columns if data.iloc[2][col] == '계' or data.iloc[2][col] == '자치구별(1)' or data.iloc[2][col] == '자치구별(2)']

# 필터링된 컬럼을 유지하고 나머지 컬럼을 삭제
data = data[columns_to_keep]

data
```

```python
# 제거할 이름 리스트
remove_names = [ '성동구', '중랑구', '강북구', '노원구', '도봉구', '은평구','양천구','강서구','구로구','영등포구','서초구','강남구','송파구','강동구']

# 2번째 열에서 이름이 remove_names에 속하지 않는 행을 선택
filtered_data = data[~data.iloc[:, 1].isin(remove_names)]
filtered_data = filtered_data.drop(2)

filtered_data
```

```python
filtered_data = filtered_data.iloc[:, 1:]

filtered_data
```

```python
# 인덱스를 재정렬
filtered_data = filtered_data.reset_index(drop=True)

filtered_data
```

```python
# 0번째 행 제거
filtered_data = filtered_data.iloc[1:]

# 1번째 행을 열 이름으로 설정
filtered_data.columns = filtered_data.iloc[0]

filtered_data = filtered_data.iloc[1:]

filtered_data
```

```python
filtered_data = filtered_data.reset_index(drop=True)

filtered_data
```

```python

# 0번째 열의 이름을 '자치구'로 변경
filtered_data = filtered_data.rename(columns={filtered_data.columns[0]: '자치구'})

filtered_data
```

```python
filtered_data = filtered_data.replace('-', 0)

filtered_data
```

```python
filtered_data.to_csv('foreign_data.csv', encoding ='utf-8-sig')
```
