```python
import pandas as pd

# 엑셀 파일 열기
file_path = '외국인부동산3.xlsx'
df = pd.read_excel(file_path)

df
```

```python
address_li = df["주소"]
```

```python
import requests

def geocode_address(address):
    api_key = 'AIzaSyBygRp967Ub0GG0DVVYOm6OOdivESjR3RQ'
    base_url = 'https://maps.googleapis.com/maps/api/geocode/json'
    
    params = {
        'address': address,
        'key': api_key,
    }
    
    response = requests.get(base_url, params=params)
    data = response.json()
    
    if data['status'] == 'OK':
        location = data['results'][0]['geometry']['location']
        latitude = location['lat']
        longitude = location['lng']
        return latitude, longitude
    else:
        return None
```

```python
# 주소를 X좌표와 Y좌표로 변환

result = []

for address in address_li:
    coordinates = geocode_address(address)
    if coordinates:
        print(f'주소 "{address}"의 지리적 좌표: {coordinates[0]}, {coordinates[1]}')
        result.append(coordinates)
    else:
        print('주소를 변환할 수 없습니다.')

print()
print(result)
```

```python
data = pd.DataFrame({'estate':df["부동산"], 'address':df["주소"], 'result':result})

data
```

```python
import folium

# 서울 지도를 중심으로 초기화
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 데이터프레임에서 주소 좌표값을 지도에 마커로 추가
for index, row in data.iterrows():
    folium.Marker(location=row['result'], popup=row['estate']).add_to(m)

# 지도를 HTML 파일로 저장
m.save('외국인부동산_지도.html')
```
