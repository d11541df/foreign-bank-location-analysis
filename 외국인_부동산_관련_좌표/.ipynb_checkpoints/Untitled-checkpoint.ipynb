```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-4,4,1204)
y = 0.25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.annotate('brackmard minimum',
             ha = 'center',
             va = 'bottom',
             xytext = (-1.5, 3.),
             xy = (0.75, -2.7),
             arrowprops = {
                 'facecolor': 'black',
                 'shrink':0.05})
plt.plot(X,y)
plt.show()
```

```python
X = np.linspace(-4,4,1204)
y1 = np.sin(X)
y2 = np.cos(X)

plt.xlabel('X')
plt.ylabel('y')

plt.plot(X,y1, c= 'k', lw = 3., 
         label ='sin(X)')
plt.plot(X,y2, c = '.5', lw = 3., ls = '--',
        label = 'cos(X)')


plt.lend(loc = 'best')
plt.show()
```
